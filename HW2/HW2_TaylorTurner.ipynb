```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("Lemonade2016-2.csv")
data.iloc[8, 0] = "7/8/2016"
data['Leaflets'] = data['Leaflets'].fillna(round(data['Leaflets'].mean(),0))
data = data.drop_duplicates()
data = data.reset_index(drop = True)
data['Sales'] = data['Lemon'] + data['Orange']
data['Revenue'] = data['Sales']*data['Price']
data.plot.scatter(x = 'Leaflets', y = 'Sales')
plt.title("Relationship Between Leaflets and Sales")
plt.show()
```

```python
plt.hist(data['Revenue'], bins = 10)
plt.title("DS201X HW2 - Revenue Histograms of Lemonade Sales")
plt.show()
```

```python
import seaborn as sns
ax = sns.boxplot(x = 'Location', y = 'Sales', hue = 'Location', data = data, palette = 'Set3')
plt.title('Sales at Park vs Beach')
plt.show()
```

```python
df_temp = data[data.Temperature >= 75]
ax = sns.boxplot(x = 'Location', y = 'Sales', hue = 'Location', data = df_temp, palette = 'Set3')
plt.title('Sales at Park vs Beach When Temp is at least 75')
plt.show()
```

```python
sns.pairplot(data, vars = ['Revenue','Temperature', 'Leaflets'], hue = 'Location', diag_kind = 'hist')
plt.show()
```
